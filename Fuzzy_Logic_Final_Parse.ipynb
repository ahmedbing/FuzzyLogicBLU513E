```python
import re
import pandas as pd
import csv

f=open("/Users/ahmedbingol/Desktop/Fuzzy Logic Proj Final/COM_Central_log.txt", "r") 

contents =f.read()
lines=re.split("[\r\n]+",contents)

Y_pattern= ('Y = (-?)(\d+\.)(\d+)')
X_pattern = ('X = (-?)(\d+\.)(\d+)')
Theta_pattern= ('Theta = (-?)(\d+\.)(\d+)')
X_Dest = ('Variable1 = (\d+)')
Y_Dest = ('Variable2 =(\d+)')
V_Pat = ('Robot_V = (-?)(\d+\.)(\d+)')
W_Pat = ('Robot_W = (-?)(\d+\.)(\d+)')

times_dict = {}
Robot_Y_Dict = {}
Robot_X_Dict = {}
Robot_Theta_Dict = {}
X_Dest_Dict = {}
Y_Dest_Dict = {}
V_Dict = {}
W_Dict = {}

line_num=len(lines)
for i in range(line_num-1):
    times_dict[i] = lines[i][11:23]
    
    p1 = re.findall(Y_pattern,lines[i])
    p2 = re.findall(X_pattern,lines[i])
    p3 = re.findall(Theta_pattern,lines[i])
    p4 = re.findall(X_Dest,lines[i])
    p5 = re.findall(Y_Dest,lines[i])
    p6 = re.findall(V_Pat,lines[i])
    p7 = re.findall(W_Pat,lines[i])
    f1=1
    f2=1
    f3=1
    f4=1
    f5=1
    f6=1
    f7=1
    
    if p7:
        W_Dict[times_dict[i]]= float(p7[0][0]+p7[0][1]+ p7[0][2])
    else:
        f7=0
    if p6:
        V_Dict[times_dict[i]]= float(p6[0][0]+p6[0][1]+ p6[0][2])
    else:
        f6=0
    if p5:
        Y_Dest_Dict[times_dict[i]]= float(p5[0])
    else:
        f5=0
    if p4:
        X_Dest_Dict[times_dict[i]]= float(p4[0])
    else:
        f4=0
    if p3:
        Robot_Theta_Dict[times_dict[i]] = float(p3[0][0]+p3[0][1]+ p3[0][2] )
    else:
        f3=0
    if p2:
        Robot_X_Dict[times_dict[i]] = float(p2[0][0]+p2[0][1]+ p2[0][2] )
    else:
        f2=0
    if p1:
        Robot_Y_Dict[times_dict[i]] = float(p1[0][0]+p1[0][1]+ p1[0][2] )
    else:
        f1=0
        
    if (f1+f2+f3+f4+f5+f6+f7 == 0):
        del times_dict[i]
```

```python
liste = []
for i in range(line_num-1):
    temp_Dic= {}
    if i in times_dict:
        
        temp_Dic['Time']=  times_dict[i]
        
        if times_dict[i] in Robot_X_Dict:
            temp_Dic['X'] = Robot_X_Dict[times_dict[i]]
        else:
            temp_Dic['X'] = 'NaN'

        if times_dict[i] in Robot_Y_Dict:
            temp_Dic['Y'] = Robot_Y_Dict[times_dict[i]]
        else:
            temp_Dic['Y'] = 'NaN'

        if times_dict[i] in Robot_Theta_Dict:
            temp_Dic['Theta'] = Robot_Theta_Dict[times_dict[i]]
        else:
            temp_Dic['Theta'] = 'NaN'

        if times_dict[i] in X_Dest_Dict:
            temp_Dic['X_Dest'] = X_Dest_Dict[times_dict[i]]
        else:
            temp_Dic['X_Dest'] = 'NaN'

        if times_dict[i] in Y_Dest_Dict:
            temp_Dic['Y_Dest'] = Y_Dest_Dict[times_dict[i]]
        else:
            temp_Dic['Y_Dest'] = 'NaN'

        if times_dict[i] in V_Dict:
            temp_Dic['Lin_Vel'] = V_Dict[times_dict[i]]
        else:
            temp_Dic['Lin_Vel'] = 'NaN'

        if times_dict[i] in W_Dict:
            temp_Dic['Ang_Vel'] = W_Dict[times_dict[i]]
        else:
            temp_Dic['Ang_Vel'] = 'NaN'
        liste.append(temp_Dic)
```

```python
Col_names= ['Time','X', 'Y', 'Theta', 'X_Dest', 'Y_Dest', 'Lin_Vel', 'Ang_Vel']
with open('Robot_Data.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=Col_names)
    writer.writeheader()
    writer.writerows(liste)
csvfile
```
